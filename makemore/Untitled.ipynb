```python
import torch
```

```python
words = open("names.txt", "r").read().splitlines()
words
```

```python
chars = sorted(set(''.join(words)))
stoi = {s:i+1 for i, s in enumerate(chars)}
itos = {i+1:s for i, s in enumerate(chars)}
stoi['.'] = 0
itos[0] = '.'
stoi, itos
```

```python
N = torch.zeros((27,27), dtype=int)
N.shape
```

```python
xs = []
ys = []
for w in words:
    S = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(S, S[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()

W = torch.randn((27,27), requires_grad=True)
num
```

```python
len(xs)
```

```python
W = torch.randn((27,27), requires_grad=True)
```

```python
import torch.nn.functional as F

#Forward
H = F.one_hot(xs, num_classes=27).float()
logits = H @ W
counts = logits.exp()
P = counts/counts.sum(1, keepdims=True)
loss = -P[torch.arange(num), ys].log().mean()
print(loss.item())
```

```python
#Backward
W.grad = None
loss.backward()

#Update
W.data += -0.5 * W.grad
```

```python
for n in range(100):
    H = F.one_hot(xs, num_classes=27).float()
    logits = H @ W
    counts = logits.exp()
    P = counts/counts.sum(1, keepdims=True)
    loss = -P[torch.arange(num), ys].log().mean()
    print(loss.item())

    #Backward
    W.grad = None
    loss.backward()

    #Update
    W.data += -50 * W.grad
```
